```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = [50000, 52964,1041, 30193]
labels = ['Annotation & Fine-tuning','Model Training Set','Hand Annotation', 'Evaluation Set']
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(8, 8))
colors = sns.color_palette("pastel")
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Dataset distribution")
plt.show()
```

```python
explode = [0, 0,0, 0.1]  # Hebt das zweite Segment hervor
plt.pie(data, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)
plt.savefig('Verteilung.png')
```
